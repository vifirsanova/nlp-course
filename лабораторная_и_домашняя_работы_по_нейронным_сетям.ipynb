```python
import random
sample_texts = random.sample(texts, 10)
for text in sample_texts:
    print(text[0:99])
```

```python
import re
cleaned_text = []
for text in sample_texts:
  if type(text) == str:
    a = re.sub(r'https?://\S+', '', text)
    b = re.sub(r'[@#]\w+', '', a)
    c = re.sub(r'\+7\s\d{3}\s\d{3}\s\d{2}\s\d{2}', '', b)
    d = re.sub(r'\+7\d{10}', '', c)
    e = re.sub(r'\W+', ' ', d) # —É–¥–∞–ª–µ–Ω–∏–µ –∑–Ω–∞–∫–æ–≤ –ø—Ä–µ–ø–∏–Ω–∞–Ω–∏—è
    f = re.sub(r'[a-zA-Z]+', ' ', e) # —É–¥–∞–ª–µ–Ω–∏–µ –ª–∞—Ç–∏–Ω—Å–∫–∏—Ö —Å–∏–º–≤–æ–ª–æ–≤
    cleaned_text.append(f)
  elif type(text) == list:
    for i in text:
      if type(i) == str:
        a = re.sub(r'https?://\S+', '', i)
        b = re.sub(r'[@#]\w+', '', a)
        c = re.sub(r'\+7\s\d{3}\s\d{3}\s\d{2}\s\d{2}', '', b)
        d = re.sub(r'\+7\d{10}', '', c)
        e = re.sub(r'\W+', ' ', d) # —É–¥–∞–ª–µ–Ω–∏–µ –∑–Ω–∞–∫–æ–≤ –ø—Ä–µ–ø–∏–Ω–∞–Ω–∏—è
        f = re.sub(r'[a-zA-Z]+', ' ', e) # —É–¥–∞–ª–µ–Ω–∏–µ –ª–∞—Ç–∏–Ω—Å–∫–∏—Ö —Å–∏–º–≤–æ–ª–æ–≤
        cleaned_text.append(f)
for text in cleaned_text:
  print(text)
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

stop_words = set(stopwords.words("russian"))
```

```python
cleaned_words = []
for text in cleaned_text:
  words = text.split()
  for word in words:
    result = word.lower()
    if result not in stop_words:
      cleaned_words.append(result)
```

```python
import spacy

# –ó–∞–≥—Ä—É–∂–∞–µ–º –º–æ–¥–µ–ª—å –¥–ª—è —Ä—É—Å—Å–∫–æ–≥–æ —è–∑—ã–∫–∞
nlp = spacy.load("ru_core_news_sm")

entities = []
for text in cleaned_words:
    doc = nlp(text)
    # –î–æ–±–∞–≤–ª—è–µ–º –º–µ—Ç–∫—É —Å—É—â–Ω–æ—Å—Ç–∏ –≤ —Å–ø–∏—Å–æ–∫
    entities.extend([ent.label_ for ent in doc.ents])
```

```python
from collections import Counter
import matplotlib.pyplot as plt

# –ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ NER
entity_counts = Counter(entities) # –ü–æ—Å—á–∏—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—Ö–æ–∂–¥–µ–Ω–∏–π –∏–º–µ–Ω–æ–≤–∞–Ω–Ω—ã—Ö —Å—É—â–Ω–æ—Å—Ç–µ–π
plt.figure(figsize=(10, 5))
plt.bar(entity_counts.keys(), entity_counts.values())
plt.xticks(rotation=45)
plt.title("–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ NER –≤ —Ç–µ–∫—Å—Ç–µ")
plt.show()
```

```python
# PoS-–∞–Ω–∞–ª–∏–∑
pos_counts = Counter()
for text in cleaned_words:
    doc = nlp(text)
    pos_counts.update([token.pos_ for token in doc])
```

```python
# –ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ PoS
plt.figure(figsize=(10, 5))
plt.bar(pos_counts.keys(), pos_counts.values())
plt.xticks(rotation=45)
plt.title("–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —á–∞—Å—Ç–µ–π —Ä–µ—á–∏")
plt.show()
```

```python
for text in cleaned_words:
  doc = nlp(text)
  for token in doc:
    a = token.lemma_
    print(token, a)
```

```python
lemmas = []
for text in cleaned_words:
  doc = nlp(text)
  for token in doc:
    a = token.lemma_
    lemmas.append(a)
print(lemmas)
```

```python
lemmas = Counter(token.lemma_ for text in cleaned_words for token in nlp(text))
print(lemmas)
```

```python
top = lemmas.most_common(20)
print(top)
```

```python
list_of_keys = [key for key, value in top]
list_of_values = [value for key, value in top]
```

```python
plt.figure(figsize=(10, 5))
plt.bar(list_of_keys, list_of_values)
plt.xticks(rotation=50, ha="right")
plt.title("–ß–∞—Å—Ç–æ—Ç–Ω–æ—Å—Ç—å –ª–µ–º–º")
plt.show()
```
